```python
import pandas as pd

licenses_df = pd.read_csv(r"..\_data\licenses_fall2022.csv")

licenses_df.head()
```

```python
import altair as alt

# Visualization 1: Breakdown of license types by status

# Prepare the data: Count the occurrences of each status within each license type
license_status_counts = licenses_df.groupby(['License Type', 'License Status']).size().reset_index(name='Count')

# Create the chart: Stacked bar chart
chart1 = alt.Chart(license_status_counts).mark_bar().encode(
    x='License Type',
    y=alt.Y('Count', stack='normalize'),
    color='License Status',
    tooltip=['License Type', 'License Status', 'Count']
).properties(
    width=600,
    height=400,
    title="Breakdown of License Types by Status"
).interactive()

chart1
```

```python
with open('../assets/json/license_type.json', 'w') as f:
    f.write(chart1.to_json())
```

```python
licenses_df['LastModifiedDate'] = pd.to_datetime(licenses_df['LastModifiedDate'])

licenses_df['YearMonth'] = licenses_df['LastModifiedDate'].dt.strftime('%Y-%m')

time_agg = licenses_df.groupby('YearMonth').size().reset_index(name='Count')

selection = alt.selection_interval(encodings=['x'])

chart2 = alt.Chart(time_agg).mark_line(point=True).encode(
    x=alt.X('YearMonth:T', axis=alt.Axis(title='Time')),
    y=alt.Y('Count:Q', axis=alt.Axis(title='Number of License Modifications')),
    tooltip=['YearMonth:T', 'Count:Q'],
    color=alt.condition(selection, alt.value('blue'), alt.value('lightgray'))
).add_params(
    selection
).properties(
    title='Trend of License Issuances and Modifications Over Time',
    width=800,
    height=400
)

chart2
```

```python
with open('../assets/json/license_time.json', 'w') as f:
    f.write(chart2.to_json())
```
